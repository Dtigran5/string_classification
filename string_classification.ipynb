```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
```

```python
df_test = pd.read_csv('test.csv')
```

```python
df_train
```

```python
df_test
```

```python
df_train = df_train.drop('id', axis = 1)
df_train = df_train.drop('keyword', axis = 1)
df_train = df_train.drop('location', axis = 1)
df_test = df_test.drop('id', axis = 1)
df_test = df_test.drop('keyword', axis = 1)
df_test = df_test.drop('location', axis = 1)
```

```python
df_train.info()
```

```python
df_train.isna().sum()
```

```python
X = df_train['text']
y = df_train['target']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
tfidf = TfidfVectorizer()
```

```python
pipe = Pipeline([('tfidf', tfidf), ('model', model)])
```

```python
pipe.fit(X, y)
```

```python
pipe = pipe.predict(df_test['text'])
```

```python
sub = pd.read_csv('sample_submission.csv')
```

```python
sub['target'] = pipe
```

```python
sub.to_csv('sub.csv', index = False, index_label = False)
```
